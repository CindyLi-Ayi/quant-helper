```python
# this multipage method is referenced from stackoverflow     

class topwindow(tk.Frame):
    def __init__(self, root, data, address,stock,interval,start,end,period):
        self.root=root
        super().__init__(root)
        self.data, self.address, self.stock, self.interval, self.start, self.end, self.period=\
                                    data,address,stock,interval,start,end,period
        self.container = tk.Frame(self,bg='lavender')
        self.container.pack(side="top", fill="both", expand=True)
        
        mainmenu=tk.Menu(self)
        datamenu=tk.Menu(mainmenu,tearoff=False)
        datamenu.add_command(label='download from yahoo',command=self.run_download)
        datamenu.add_command(label='upload',command=self.run_upload)
        mainmenu.add_cascade(label='data',menu=datamenu)
        
        edamenu=tk.Menu(mainmenu,tearoff=False)
        edamenu.add_command(label='Nan Value',command=self.run_nan)
        edamenu.add_command(label='distribution',command=self.run_dist)
        edamenu.add_command(label='trend',command=self.run_trend)
        edamenu.add_command(label='correlation',command=self.run_corr)
        mainmenu.add_cascade(label='EDA',menu=edamenu)
        
#         indicatormenu=tk.Menu(mainmenu,tearoff=False)
#         backtestmenu=tk.Menu(indicatormenu,tearoff=False)
#         m1=tk.Menu(featuremenu,tearoff=False)
#         for f in ['Simple Moving Average','Weighted Moving Average',\
#             'Exponential Moving Average','Bollinger Bands']:
#             m1.add_command(label=f,command=lambda : self.run_indicator(f))
#         backtestmenu.add_cascade(label="Overlap Studies",menu=m1)
#         m2=tk.Menu(featuremenu,tearoff=False)
#         for f in ['Average Directional Movement Index','Aroon Oscillator',\
#                 'Commodity Channel Index','Moving Average Convergence/Divergence',\
#                 'Absolute Price Oscillator','Percentage Price Oscillator',\
#                 'Balance Of Power','Relative Strength Index','Momentum']:
#             m2.add_command(label=f,command=lambda : self.run_indicator(f))
#         backtestmenu.add_cascade(label="Momentum Indicator",menu=m2)
#         m3=tk.Menu(featuremenu,tearoff=False)
#         for f in ['True Range','Average True Range','Normalized Average True Range']:
#             m3.add_command(label=f,command=lambda : self.run_indicator(f))
#         backtestmenu.add_cascade(label="Volatility Indicator",menu=m3)
#         m4=tk.Menu(featuremenu,tearoff=False)
#         for f in ['Chaikin A/D Line','Chaikin A/D Oscillator','On Balance Volume']:
#             m4.add_command(label=f,command=lambda : self.run_indicator(f))
#         backtestmenu.add_cascade(label="Volume Indicator",menu=m4)
#         indicatormenu.add_cascade(label='Backtest',menu=backtestmenu)
#         indicatormenu.add_command(label='performance',command=self.run_performance)
#         mainmenu.add_cascade(label='Indicator',menu=indicatormenu)
        
#         mlmenu=tk.Menu(mainmenu,tearoff=False)
#         featuremenu=tk.Menu(mlmenu,tearoff=False)
#         m1=tk.Menu(featuremenu,tearoff=False)
#         for f in ['Simple Moving Average','Weighted Moving Average',\
#             'Exponential Moving Average','Bollinger Bands']:
#             m1.add_command(label=f,command=lambda : self.run_feature(f))
#         featuremenu.add_cascade(label="Overlap Studies",menu=m1)
#         m2=tk.Menu(featuremenu,tearoff=False)
#         for f in ['Average Directional Movement Index','Aroon Oscillator',\
#                 'Commodity Channel Index','Moving Average Convergence/Divergence',\
#                 'Absolute Price Oscillator','Percentage Price Oscillator',\
#                 'Balance Of Power','Relative Strength Index','Momentum']:
#             m2.add_command(label=f,command=lambda : self.run_feature(f))
#         featuremenu.add_cascade(label="Momentum Indicator",menu=m2)
#         m3=tk.Menu(featuremenu,tearoff=False)
#         for f in ['True Range','Average True Range','Normalized Average True Range']:
#             m3.add_command(label=f,command=lambda : self.run_feature(f))
#         featuremenu.add_cascade(label="Volatility Indicator",menu=m3)
#         m4=tk.Menu(featuremenu,tearoff=False)
#         for f in ['Chaikin A/D Line','Chaikin A/D Oscillator','On Balance Volume']:
#             m4.add_command(label=f,command=lambda : self.run_feature(f))
#         featuremenu.add_cascade(label="Volume Indicator",menu=m4)
#         mlmenu.add_cascade(label='feature',menu=featuremenu)
#         modelmenu=tk.Menu(mlmenu,tearoff=False)
#         for f in ['linear regression']:
#             modelmenu.add_command(label=f,command=lambda : self.run_feature(f))
#         mlmenu.add_cascade(label='model',menu=modelmenu)
#         mainmenu.add_cascade(label='ML',menu=mlmenu)
        
        root.config(menu=mainmenu)
    def run_download(self):
        download=download_page(self, self.data, self.stock,self.interval,self.start,self.end,self.period)
        download.place(in_=self.container, x=0, y=0, relwidth=1, relheight=1)
        download.lift()
        

    def run_upload(self):
        upload=upload_page(self, data, address)
        upload.place(in_=self.container, x=0, y=0, relwidth=1, relheight=1)
        upload.lift()
        self.root.bind('<Return>',upload.uploading)

    def run_nan(self):
        nan=nan_page(self, data.data)
        nan.place(in_=self.container, x=0, y=0, relwidth=1, relheight=1)
        nan.lift()
        
    def run_dist(self):
        dist=dist_page(self, data.data)
        dist.place(in_=self.container, x=0, y=0, relwidth=1, relheight=1)
        dist.lift()
        
    def run_trend(self):
        trend=trend_page(self, data.data)
        trend.place(in_=self.container, x=0, y=0, relwidth=1, relheight=1)
        trend.lift()
    
    def run_corr(self):
        corr=corr_page(self, data.data)
        corr.place(in_=self.container, x=0, y=0, relwidth=1, relheight=1)
        corr.lift()
        self.root.bind('<Return>',corr.onselection)
    
    def run_indicator(self,func_name):
        indicator=indicator_page(self,data.data,data.stock,func_name)
        indicator.place(in_=self.container, x=0, y=0, relwidth=1, relheight=1)
        indicator.lift()
    
    def run_feature(self,func_name):
        feature=feature_page(self,data.data,data.stock,func_name)
        feature.place(in_=self.container, x=0, y=0, relwidth=1, relheight=1)
        feature.lift()
```

```python
root = tk.Tk()
root.title('Quant Helper')
address,stock,interval,start,end,period=\
                tk.StringVar(),tk.StringVar(),tk.StringVar(),tk.StringVar(),tk.StringVar(),tk.StringVar()
data=ydata()

main = topwindow(root,data,address,stock,interval,start,end,period)
main.pack(side="top", fill="both", expand=True)
root.geometry("1440x900")
root.mainloop()
```
